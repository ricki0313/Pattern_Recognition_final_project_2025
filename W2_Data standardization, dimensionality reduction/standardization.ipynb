```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Users\use\Desktop\NCCU_STAT\Pattern Recognition\final\PR_online retail\W1_Data cleaning, aggregation, visualization\customer_features.csv")
```

```python
df.head(3)
```

```python
features = ["Recency","Frequency","Monetary","AvgUnitPrice","UniqueItems","AvgQuantPerOrder"]
X = df[features]
```

```python
scalar = StandardScaler()
Z = scalar.fit_transform(X)
```

```python
print(Z)
```

```python
df_scaled = pd.DataFrame(Z, columns=[f"{col}_scaled" for col in features])
```

```python
df_scaled = pd.concat([df[["Customer ID"]], df_scaled], axis=1)
df_scaled.head(3)
```

```python
df_scaled.describe()
```

```python
df_scaled.to_csv("scaled_customer_features.csv", index=False)
```
