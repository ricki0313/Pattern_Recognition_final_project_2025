<h1> Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/customer_features.csv")
df.head(3)
```

<h3> Visualize the data of each feature

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Recency"], bins=30, color="gray")
plt.xlabel("Recency (days)")
plt.ylabel("Count")
plt.title("Recency Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Frequency"], bins=30, color="gray")
plt.xlabel("Frequency (number of purchases)")
plt.ylabel("Count")
plt.title("Frequency Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Monetary"], bins=30, color="gray")
plt.xlabel("Monetary (Total spending)")
plt.ylabel("Count")
plt.title("Monetary Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["AvgQuantPerOrder"], bins=30, color="gray")
plt.xlabel("AvgQuantPerOrder (avg quantity per invoice)")
plt.ylabel("Count")
plt.title("AvgQuantPerOrder Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["UniqueItems"], bins=30, color="gray")
plt.xlabel("UniqueItems (distinct products purchased)")
plt.ylabel("Count")
plt.title("UniqueItems Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["AvgUnitPrice"], bins=30, color="gray")
plt.xlabel("AvgUnitPrice")
plt.ylabel("Count")
plt.title("AvgUnitPrice Distribution")
plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="coolwarm", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right')
plt.yticks(range(len(corr.columns)), corr.columns)

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        value = corr.iloc[i, j]
        plt.text(j, i, f"{value:.2f}", ha='center', va='center', color="black", fontsize=8)

plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
corr
```
