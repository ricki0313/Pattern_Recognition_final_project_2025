<h1> Data Cleaning

```python
import pandas
```

```python
# This notebook expects the raw dataset to be located at:
# data/raw/online_retail_II.xlsx
df = pandas.read_excel("../data/raw/online_retail_II.xlsx", sheet_name="Year 2010-2011")
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Customer ID'])
```

```python
df['Invoice'].apply(type).value_counts()
```

```python
df["Invoice"] = df["Invoice"].astype(str)
```

```python
df[df['Invoice'].str.startswith('C')].head()
```

```python
df = df[~df['Invoice'].str.startswith('C')]
df.shape
```

```python
(df['Quantity'] < 0).sum()
```

```python
(df['Price'] < 0).sum()
```

```python
df['Customer ID'].isna().sum()
```

```python
df.info()
```

```python
df['StockCode'] = df['StockCode'].astype(str)
df['Description'] = df['Description'].astype(str)
df['Country'] = df['Country'].astype(str)
```
